```julia
using Compose
```

```julia
compose(context(), fill("tomato"),
        (context(0.0, 0.0, 0.5, 0.5), circle()),
        (context(0.5, 0.5, 0.5, 0.5), circle()))
```

```julia
compose(context(),
        (context(units=UnitBox(0, 0, 1000, 1000)),
         polygon([(0, 1000), (500, 1000), (500, 0)]),
         fill("tomato")),
        (context(),
         polygon([(1, 1), (0.5, 1), (0.5, 0)]),
         fill("bisque")))
```

```julia
using Colors

product =
    compose(
            context(),
            (context(), line([(0.09, 0.91), (0.91, 0.09)]), stroke(colorant"black"), linewidth(1.5mm)),
            (context(), line([(0.09, 0.09), (0.91, 0.91)]), stroke(colorant"black"), linewidth(1.5mm)),
            (context(), circle(), stroke(colorant"black"), fill(nothing), linewidth(1.5mm))
)
```

```julia
sum =
    compose(
            context(),
            (context(), line([(0.5, 1.0), (0.5, 0.0)]), stroke(colorant"black"), linewidth(1.5mm)),
            (context(), line([(1., 0.5), (0.0, 0.5)]), stroke(colorant"black"), linewidth(1.5mm)),
            (context(), circle(), stroke(colorant"black"), fill(nothing), linewidth(1.5mm))
)
```
